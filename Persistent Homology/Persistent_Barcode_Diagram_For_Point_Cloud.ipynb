# Intro
Generating persistence barcode and diagram figures for a point cloud.

```python
from tda_utils import*
```

```python
pt_cloud = np.load('Datasets/point_cloud_in_plane.npy')
```

```python
fig = plt.figure(figsize = (10,10))
plt.scatter(pt_cloud[:,0],pt_cloud[:,1])
plt.title('Point Cloud',fontsize = 20)
plt.savefig('Figures/Point_Cloud.png',bbox_inches = 'tight')
plt.axis('equal')
plt.show()
```

```python
rips_complex = gd.RipsComplex(points = pt_cloud,max_edge_length = 20)
simplex_tree = rips_complex.create_simplex_tree(max_dimension = 2)
dgm = simplex_tree.persistence()
```

```python
fig = plt.figure(figsize = (25,7))
ax = fig.add_subplot()
plot_barcode(dgm,r =3,ax = ax)
ax.set_title('Persistence Barcode',fontsize = 28)
ax.set_xlabel(f'Filtration Value $r$',fontsize = 18)
plt.savefig('Figures/persistent_barcode_pt_cloud.png',bbox_inches = 'tight')
plt.show()
```

```python
fig = plt.figure(figsize = (25,7))
ax = fig.add_subplot()
plot_persistence_diagram(dgm,r =3,ax = ax)
ax.set_title('Persistence Diagram',fontsize = 20)
plt.savefig('Figures/persistent_diagram_pt_cloud.png',bbox_inches = 'tight')
plt.show()
```

```python
fig = plt.figure(figsize = (5,5))
ax = fig.add_subplot()
r_val = 2.6
plot_VR(pt_cloud, r = r_val,ax= ax)
ax.set_title(f'VR - Complex with $r$ = {r_val}',fontsize = 18)
plt.savefig(f'Figures/vr_complex_r_{r_val}.png',bbox_inches = 'tight')
plt.show()
```
